```python
import csv
import pandas as pd
import numpy as np
```

```python
# Here, we create list, which will act as the Headers to 

headerList = []
for i in range(58):
    headerList.append("kp" + str(i+1))

headerList.append('output')
print(headerList)
```

```python
# Here we need to be careful --> we create an absolute new 'csv' data
# Here we specify Headers and write to the csv 

with open("keypoints_data.csv", 'w') as file:
    dw = csv.DictWriter(file, delimiter=',', 
                        fieldnames=headerList)
    dw.writeheader()
  
# display csv file
fileContent = pd.read_csv("keypoints_data.csv")
```
