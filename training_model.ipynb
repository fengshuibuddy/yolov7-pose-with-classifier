```python
import pandas as pd
import csv
import numpy as np
```

## This part  --->  is when we already have the ready-to-train data 

```python
data = pd.read_csv('keypoints_data.csv')
```

```python
data
```

```python
# Все как обычно

# Initialize 'X' data
X = data.iloc[:,:-1]

# Initialize 'y' data
y = data.iloc[:, -1:]
```

```python
X
```

```python
y
```

```python
# Типичная подготовка Даты идет 

from sklearn.model_selection import train_test_split 

# Split the data. 90% training data, 10% test data
X_train, X_test, yy_train, yy_test = train_test_split(X, y, test_size=0.15, shuffle=True)

print("X_train shape: ", X_train.shape)
print("X_test shape: ", X_test.shape)
print("yy_train shape: ", yy_train.shape)
print("yt_test shape: ", yy_test.shape)
```

```python
# Здесь я пока обычную модель выбрал -- а так можно и другие даты тоже

from sklearn import tree

clf = tree.DecisionTreeClassifier()
```

```python
# Train the model

clf.fit(X_train, yy_train)
```

```python
# Run model 

y_pred = clf.predict(X_test)
print((y_pred))
```

```python
print(yy_test['output'])
```

```python
print(y_pred)
```

```python
# Saving model
import pickle

filename = 'decision_tree_model.sav'
pickle.dump(clf, open(filename, 'wb'))
```
